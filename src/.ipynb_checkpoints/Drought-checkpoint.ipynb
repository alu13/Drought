<h1>Part 2: Creating a basic model of farmer competition and supply and demand </h1>

In this section I will use agricultural output to explain how much rainfall farms need. I will create a semi-complex competiting species model for crops like cotton and almonds to describe how they get impacted by the presence of other water-using crops. This is essentially a time-lagged supply and demand model that we covered in class mixed with a little bit of competation between difficult products. The farmers are the suppliers and demand is covered by a basic model of

Demand = base_price * ratio_of_farmers_that_make_crop

Each farmer is capable of growing one of five crops. Cotton, figs, almonds, tomatoes, and grapes. These are five of the most commonly grown crops in Fresno and they each represent a different price and water requirement combination. 

The model will work like this. Every year, each farmer will grow their crop and bring it to market. Every year, the demand will change based on what was brought to market and the farmers will adapt their supply based on the demand. In this simple process, the model will converge on a steady state where the demand (in $) of all five crops is similar due to the ratio of the supply.

There are a couple of elements of randomness.

1. Each farmer is given a random crop
2. Each farmer is given a random "openness to change" variable

These forms of randomness are to ensure that the model is robust under noise. They are NOT reflected by real world data. The real world is significantly more complex.

The price and water usage of each crop is shown below.

<h3> Setup </h3>

The correlation is clear. Crops with a higher water cost are more profitable. And this bleeds into the complexity of agriculture. Water is very inexpensive (because it's a natural resource), and farmers have access to significant amounts of water at significantly cheaper prices. At the most basic level, water is a CPR, or a common pool resource, and farmers can profit from taking more.

It's important to note that almonds are a huge outlier in this dataset because they are both very profitable and very water-hungry.

Below is a basic model of the situation where water price is zero.

<h3> Model Code </h3>


```python
from Crop import crop
from Farmer import farmer
%load_ext autoreload
%autoreload 2

#Hyperparameters. Can set to whatever
num_farmers = 100
years = 20
empty_dict = {
    "cotton": 0,
    "figs": 0,
    "almonds": 0,
    "tomatoes": 0,
    "grapes": 0
}

#Setting crop and farmer tags
crop = crop()
list_farmers = []
for i in range(num_farmers):
    list_farmers.append(farmer())



def predict(farmers, crops, detail = False):
    
    #Instantiating empty arrays. Eventaully, the number of farmer per year that grow the crop with populate them    
    cotton = []
    figs = []
    almonds = []
    tomatoes = []
    grapes = []
    
    #Looping every year
    for i in range(years):
        
        #Filling crop arrays
        year_crops = empty_dict.copy()
        for j in range(num_farmers):
            farmer_crop = farmers[j].crop
            year_crops[farmer_crop] += 1
    #     print(year_crops)
        cotton.append(year_crops['cotton'])
        figs.append(year_crops['figs'])
        almonds.append(year_crops['almonds'])
        tomatoes.append(year_crops['tomatoes'])
        grapes.append(year_crops['grapes'])
    
        #Updating crop prices
        new_prices = crops.update(year_crops)
        
        #Updating farmers
        vals = list(new_prices.values())
        keys = list(new_prices.keys())
        max_price = max(vals)
        max_crop = keys[vals.index(max_price)]
        for j in range(num_farmers):
            farmer_crop_price = new_prices[farmers[j].crop]
            farmers[j].update(max_price, max_crop, farmer_crop_price)
        
        #Checking for enumerate
        if i == (years - 1):
            if detail == True:
                print("crop prices = " + str(new_prices))
    return cotton, figs, almonds, tomatoes, grapes

def plot(cotton, figs, almonds, tomatoes, grapes):
    
    width = .5
    labels = np.arange(0, len(cotton))
    fig, ax = plt.subplots()
    
    break1 = np.add(cotton, figs).tolist()
    break2 = np.add(break1, almonds).tolist()
    break3 = np.add(break2, tomatoes).tolist()
    
    ax.bar(labels, cotton, width, label='cotton', color = 'red')
    ax.bar(labels, figs, width, label = 'figs', bottom = cotton, color = 'green', edgecolor='white')
    ax.bar(labels, almonds, width, label='almonds', bottom = break1, color = 'purple', edgecolor='white')
    ax.bar(labels, tomatoes, width, label='tomatoes', bottom = break2, color = 'black', edgecolor='white')
    ax.bar(labels, grapes, width, label='grapes', bottom = break3, color = 'blue', edgecolor='white')

    ax.set_ylabel('# of farmers')
    ax.set_title('Farmers per crop')
    ax.legend()

    plt.show()

cotton, figs, almonds, tomatoes, grapes = predict(list_farmers, crop)
plot(cotton, figs, almonds, tomatoes, grapes)
```

This model does a good job of representing the steady state where the ratio between #farmers per crop is the ratio between the prices per crop which is the expected steady state of this. However, this model already shows the common pool resource problem. Too many people are greedy and want to sell almonds because they are the most expensive, which drains already strained resources.

However, the change is a little much for farmers. In the following cell, I increased farmer resilience, which leads to a steadier steady state over time.

```python
from Crop import crop

crop = crop()
list_farmers = []
for i in range(num_farmers):
    list_farmers.append(farmer(200))
cotton, figs, almonds, tomatoes, grapes = predict(list_farmers, crop, detail = True)
plot(cotton, figs, almonds, tomatoes, grapes)
```

This does a better job of representing the trend towards the ratio steady state and the steadiness of the steady state. Almonds are still the most profitable, but not profitable enough to make the other farmers change professions. Almonds still dominate though.

<h3> Groundwater plot </h3>

```python
em_water_supply = 2000000
farmers = np.array([5, 5, 5, 5, 5])
yearly_water = np.sum(crops['water'] * farmers)
yearly_water_graph = np.array([])
yearly_water_supply = np.array([])
print(yearly_water)
while(em_water_supply > 0):
    farmers = np.array([0, 0, np.sum(farmers), 0, 0])
    farmer_water = np.sum(crops['water'] * farmers)
    yearly_water_graph = np.append(yearly_water_graph, [yearly_water])
    em_water_supply += yearly_water - farmer_water
    yearly_water_supply = np.append(yearly_water_supply, em_water_supply)
plt.plot(np.arange(0, 8), yearly_water_supply)
```

This isn't a particularly complex graph but it represents the problem of common pool resources. Some things just aren't really free. Nobody gets any water and the common pool runs out over time. <h3> Varied Water Prices Code </h3>

```python
from Crop import crop
def pow_0(years):
    return np.random.randint(5, 30, years) / 100
        
def predict_with_variance(farmers, crops, years, pri_wat, detail = False):
    cotton = []
    figs = []
    almonds = []
    tomatoes = []
    grapes = []
    for i in range(years):
        year_crops = empty_dict.copy()
        for j in range(num_farmers):
            farmer_crop = farmers[j].crop
            year_crops[farmer_crop] += 1
    #     print(year_crops)
        cotton.append(year_crops['cotton'])
        figs.append(year_crops['figs'])
        almonds.append(year_crops['almonds'])
        tomatoes.append(year_crops['tomatoes'])
        grapes.append(year_crops['grapes'])
        price_of_water = pri_wat[i]
            
        print("pow = " + str(price_of_water))
        new_prices = crops.update(year_crops, price_of_water)
        vals = list(new_prices.values())
        keys = list(new_prices.keys())
        max_price = max(vals)
        max_crop = keys[vals.index(max_price)]
        for j in range(num_farmers):
            farmer_crop_price = new_prices[farmers[j].crop]
            farmers[j].update(max_price, max_crop, farmer_crop_price)
        if i == (years - 1):
            if detail == True:
                print("crop prices = " + str(new_prices))
    return cotton, figs, almonds, tomatoes, grapes

years = 40
crop = crop(.18)
list_farmers = []
for i in range(num_farmers):
    list_farmers.append(farmer(200))
pri_wat = pow_0(years)
cotton, figs, almonds, tomatoes, grapes = predict_with_variance(list_farmers, crop, years, pri_wat, detail = True)
plot(cotton, figs, almonds, tomatoes, grapes)
```

Again, it's important to remember that farming turns to complete trash at water costs that are 1/30th of the urban population. The resulting values are pretty chaotic, but they make sense with regards to varied water pricing. Almonds and cotton take over grapes and tomatoes as the prices for water decrease. As a whole, this looks relatively sustainable. However, again, this isn't tied to drought.

<h3> Groundwater storage </h3>

The whole point of this was to demonstrate that if water isn't valued, farmers would abuse this common pool resource and all grow almonds. We must analyze how the groundwater (backup water storage) diminishes over time to set accurate prices on water. If you would like to read up on how drought affects groundwater depletion, the link below is very helpful.

https://www.sciencemag.org/news/2020/04/droughts-exposed-california-s-thirst-groundwater-now-state-hopes-refill-its-aquifers

Say our groundwater is capable of sustaining all of agriculture without precipitation for 5 years. And say Fresno needs around 13 inches of rainwater per day (It needs more but it gets less) to sustain an an even distribution of agriculture.

<h3> Setup </h3>

```python
yearly_water = 20 * np.sum(crops['water'].to_numpy())
rain_inch = yearly_water / 13
groundwater = 5 * yearly_water
```

<h3> Groundwater Code </h3>

```python
def predict_with_groundwater(farmers, crops2, years, pri_wat, prcp, start_ground = groundwater, detail = False):
    #Instantiation
    ground = start_ground
    cotton = []
    figs = []
    almonds = []
    tomatoes = []
    grapes = []
    groundwaters = []
    
    #Loop over iterations (yeras)
    for i in range(years):
        year_crops = empty_dict.copy()
        
        #Creating distribution of farmer crops
        for j in range(num_farmers):
            farmer_crop = farmers[j].crop
            year_crops[farmer_crop] += 1
    #     print(year_crops)
    
        cotton.append(year_crops['cotton'])
        figs.append(year_crops['figs'])
        almonds.append(year_crops['almonds'])
        tomatoes.append(year_crops['tomatoes'])
        grapes.append(year_crops['grapes'])
        
        #Updating groundwater
        price_of_water = pri_wat[i]
        yearly_rain = prcp[i]
#         print(yearly_rain)
        ground += yearly_rain * rain_inch
#         print(year_crops.values())
#         print(crops['water'].to_numpy())
        ground -= np.sum(list(year_crops.values()) * crops['water'].to_numpy())
#         print("ground = " + str(ground))
        groundwaters.append(ground)
        
        #Updating crops
        new_prices = crops2.update(year_crops, price_of_water)
        
        #Updating farmers
        vals = list(new_prices.values())
        keys = list(new_prices.keys())
        max_price = max(vals)
        max_crop = keys[vals.index(max_price)]
        for j in range(num_farmers):
            farmer_crop_price = new_prices[farmers[j].crop]
            farmers[j].update(max_price, max_crop, farmer_crop_price)

        #IF to print crop prices in the last year
        if i == (years - 1):
            if detail == True:
                print("crop prices = " + str(new_prices))
    return cotton, figs, almonds, tomatoes, grapes, groundwaters

def plot_ground(groundwaters):
    width = .5
    labels = np.arange(0, len(groundwaters))
    fig, ax = plt.subplots()
    
    
    ax.bar(labels, groundwaters, width, label='groundwater', color = 'red')
    ax.set_ylabel('water (meters cubed)')
    ax.set_title('Groundwater storage')
    ax.legend()

    plt.show()
```

```python
from Crop import crop

prcp = fresno_temp_prec['PRCP'].to_numpy()
max_pri_wat = .45
adjusted_prcp = max_pri_wat * (prcp_min / prcp)
pri_wat = adjusted_prcp[len(prcp) - years - 1:]

years = 40
crop = crop(.18)
list_farmers = []
for i in range(num_farmers):
    list_farmers.append(farmer(150))
yearly_prcp = prcp[len(prcp) - years - 1:]
print("average rainfall = " + str(np.mean(yearly_prcp)))

cotton, figs, almonds, tomatoes, grapes, groundwaters = predict_with_groundwater(list_farmers, crop, years, pri_wat, yearly_prcp, start_ground = groundwater, detail = True)
plot(cotton, figs, almonds, tomatoes, grapes)
plot_ground(groundwaters)
```

Our groundwater storage slowly disappears with a max price of .45 because our our extreme addiction to the profitability of almonds. We have an obscene addiction to almonds because they're so profitable which mutilates our groundwater supply even with a doubled price of water. This actually passes a validity test because this is reflected in the real world. The growth of orchards and vineyards due to profitability and low cost of water are destroying California's groundwater supply.

<h3> Efficient farming </h3>

```python
def predict_with_groundwater_efficient(farmers, crops2, years, pri_wat, prcp, start_ground = groundwater, detail = False):
    #Instantiation
    ground = start_ground
    cotton = []
    figs = []
    almonds = []
    tomatoes = []
    grapes = []
    groundwaters = []
    
    #Loop over iterations (years)
    for i in range(years):
        year_crops = empty_dict.copy()
        
        #Creating distribution of farmer crops
        for j in range(num_farmers):
            farmer_crop = farmers[j].crop
            year_crops[farmer_crop] += 1
    #     print(year_crops)
    
        cotton.append(year_crops['cotton'])
        figs.append(year_crops['figs'])
        almonds.append(year_crops['almonds'])
        tomatoes.append(year_crops['tomatoes'])
        grapes.append(year_crops['grapes'])
        
        #Updating groundwater
        price_of_water = pri_wat[i] / 2 #Div by two because twice as efficient
        yearly_rain = prcp[i]
#         print(yearly_rain)
        ground += yearly_rain * rain_inch
#         print(year_crops.values())
#         print(crops['water'].to_numpy())

        ground -= np.sum(list(year_crops.values()) * crops['water'].to_numpy()) / 2 #Accopmlishes same goal as div crops['water'] by 2
#         print("ground = " + str(ground))
        groundwaters.append(ground)
        
        #Updating crops
        new_prices = crops2.update(year_crops, price_of_water)
        
        #Updating farmers
        vals = list(new_prices.values())
        keys = list(new_prices.keys())
        max_price = max(vals)
        max_crop = keys[vals.index(max_price)]
        for j in range(num_farmers):
            farmer_crop_price = new_prices[farmers[j].crop]
            farmers[j].update(max_price, max_crop, farmer_crop_price)

        #IF to print crop prices in the last year
        if i == (years - 1):
            if detail == True:
                print("crop prices = " + str(new_prices))
    return cotton, figs, almonds, tomatoes, grapes, groundwaters
```

Water sustainability is an enormous problem in California due to erratic precipitation (as shown in part 1) and the profitability of water-heavy crops (as shown in part 2). This model has shown a number of effective techniques. The most effective technique is raising the price of water to match drought so that water-heavy crops don't dominate the rest.
